```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import FormatStrFormatter
label_size = 20
mpl.rcParams['xtick.labelsize'] = 15*3
mpl.rcParams['ytick.labelsize'] = 15*3
```

```python
toy3_ls = np.load("front/Toy_3_LS.npy")
toy3_epo = np.load("front/Toy_3_EPO.npy")
toy3_MH = np.load("front/Toy_3_MH.npy")
toy3_cosmos = np.load("front/Toy_3_COSMOS.npy")
toy3_truth = np.load("front/toy3_truth.npy")
toy2_ls = np.load("front/Toy_2_LS.npy")
toy2_epo = np.load("front/Toy_2_EPO.npy")
toy2_MH = np.load("front/Toy_2_MH.npy")

toy2_cosmos = np.load("front/Toy_2_COSMOS.npy")
toy2_truth = np.load("front/Toy_2_truth.npy")
toy1_ls = np.load("front/Toy_1_LS.npy")
toy1_epo = np.load("front/Toy_1_EPO.npy")
toy1_MH = np.load("front/Toy_1_MH.npy")
toy1_cosmos = np.load("front/Toy_1_COSMOS.npy")
toy1_truth = np.load("front/Toy_1_truth.npy")
```

```python
toy4_ls = np.load("front/Toy_4_LS.npy")
toy4_epo = np.load("front/Toy_4_EPO.npy")
toy4_MH = np.load("front/Toy_4_MH.npy")
toy4_cosmos = np.load("front/Toy_4_COSMOS.npy")
toy4_truth = np.load("front/Toy_4_truth.npy")
```

```python
u = np.linspace(0, np.pi/2, 200)
v = np.linspace(0, np.pi/2, 200)
r = 1
x = r*np.outer(np.cos(u), np.sin(v))
y = r*np.outer(np.sin(u), np.sin(v))
z = r*np.outer(np.ones(np.size(u)), np.cos(v))
```

```python
fig = plt.figure()
fig.set_figheight(5*3)
fig.set_figwidth((20/3)*4)

toy1 = [toy1_ls, toy1_cosmos, toy1_epo, toy1_MH]
for i in range(1, 5):
    plt.subplot(3, 4, i+4)
    plt.plot(toy1_truth[:, 0], toy1_truth[:, 1], color='red')
    plt.scatter(toy1[i-1][:, 0], toy1[i-1][:, 1], s=20, color='blue')
    if i == 1:
        plt.ylabel("Problem 1", fontsize=label_size)

fig.tight_layout()

toy2 = [toy2_ls, toy2_cosmos, toy2_epo, toy2_MH]
for i in range(1, 5):
    plt.subplot(3, 4, i)
    plt.plot(toy2_truth[:, 0], toy2_truth[:, 1], color='red')
    plt.scatter(toy2[i-1][:, 0], toy2[i-1][:, 1], s=20, color='blue')
    if i == 1:
        plt.ylabel("Problem 2", fontsize=label_size)

fig.tight_layout()

toy3 = [toy3_ls, toy3_cosmos, toy3_epo, toy3_MH]
name = ["PHN-LS", "COSMOS", "PHN-EPO", 'PHN-HVI']
for i in range(1, 5):
    plt.subplot(3, 4, i+8)
    plt.plot(toy3_truth[:, 0], toy3_truth[:, 1], color='red')
    plt.scatter(toy3[i-1][:, 0], toy3[i-1][:, 1], s=20, color='blue')
    plt.xlabel(name[i-1], fontsize=label_size)
    if i == 1:
        plt.ylabel("Problem 3", fontsize=label_size)

fig.tight_layout()

plt.savefig('front/Toy_Plots.pdf')
```
